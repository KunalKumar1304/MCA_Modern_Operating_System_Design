```python
from collections import deque

class Container:
    def __init__(self, name, cpu_time):
        self.name = name
        self.cpu_time = cpu_time  # total CPU time required
    def __str__(self):
        return f"{self.name} (Remaining: {self.cpu_time})"
def container_scheduler(containers, time_quantum):
    queue = deque(containers)
    time = 0

    print("---- Container Scheduling Simulation ----")
    while queue:
        container = queue.popleft()

        print(f"\nTime {time}: Running {container.name}")

        if container.cpu_time > time_quantum:
            time += time_quantum
            container.cpu_time -= time_quantum
            print(f"{container.name} executed for {time_quantum} units")
            queue.append(container)
        else:
            time += container.cpu_time
            print(f"{container.name} executed for {container.cpu_time} units and finished")
            container.cpu_time = 0

    print("\nAll containers have completed execution.")
    print("Total CPU Time:", time)


# -------- Example --------
containers = [
    Container("Container-A", 10),
    Container("Container-B", 5),
    Container("Container-C", 8)
]

time_quantum = 3
container_scheduler(containers, time_quantum)
```
